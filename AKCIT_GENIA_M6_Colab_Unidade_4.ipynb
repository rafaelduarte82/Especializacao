## Tratamento de Dados Faltantes e Duplicados:

* **Idades:** as idades faltantes so substitudas pela mdia das idades existentes, o que  uma prtica comum para manter a distribuio geral.

```python
# Mostrando dados faltantes na coluna 'age' (idades)
titanic[titanic['age'].isnull()]
```

# Anlise de Correlaes e Propores

Nesta seo, analisaremos as correlaes entre as variveis do *dataset* Titanic e as propores significativas que podem indicar fatores influentes na sobrevivncia dos passageiros. Utilizaremos tcnicas de visualizao para ilustrar essas relaes, facilitando a identificao de padres e *insights*.


## Calcular Correlaes

Uma das ferramentas mais utilizadas para visualizao de correlaes entre variveis  o grfico de calor. Este grfico  apresentado em forma de matriz quadrada, em que cada linha, e suas respectivas colunas, representam uma varivel. Assim, a coluna 0 representa a mesma varivel que a linha 0 (`survived` no exemplo abaixo). O valor apresentado nas clulas da matriz diz a correlao entre as variveis em questo. Assim,  correto afirmar que o triangular inferior  igual ao triangular superior da matriz. O grfico de calor  interessante, pois facilita a visualizao utilizando cores mais escuras para correlaes diretas mais fortes e cores mais claras para correlaes indiretas mais fortes.

No grfico a seguir,  possvel observar nas clulas `5x0` e `0x5` o valor `0.25`, que demonstra uma correlao importante entre a taxa paga (`fare`) e a sobrevivncia do passageiro (`survived`). Uma correlao ainda mais expressiva  apresentada nas clulas `1x0` e `0x1`, com o valor `-0.34`. Este valor indica uma correlao inversa entre a classe e a sobrevivncia do passageiro, ou seja, passageiros de classes menores (primeira classe, por exemplo) tiveram maior sobrevivncia. A maior correlao apresentada no conjunto de dados, todavia,  inversa entre `sibsp` e `alone`(`-0,61`), indicando que pessoas com irmos/cnjuges a bordo do Titanic no estavam viajando sozinhas.

```python
# Somente atributos numricos so considerados
plt.suptitle("Grfico de Calor das Correlaes entre os Atributos Numricos")
numericos = titanic.select_dtypes(include=['int', 'float', 'bool'])
sns.heatmap(numericos.corr(), annot=True, cmap='Blues')
```

## Anlisar Propores

### Homens x Mulheres

Os resultados a seguir mostram que apenas 21,72% dos homens sobreviveram. No caso das mulheres, 73,97% sobreviveram. Isso indica um reflexo da abordagem costumeira da poca de salvar mulheres primeiro.

Nos cdigos, inicialmente filtramos (com `loc`) o *DataFrame* pelo gnero (`titanic['sex'] == 'male'` ou `titanic['sex'] == 'female'`), selecionando apenas a coluna `survived`. Em seguida, contamos (`count()`) o total de pessoas daquele gnero e os totais para cada valor em `survived` (`value_counts()`). Por fim, as porcentagens so calculadas.

```python
# Filtrando apenas a coluna 'survived' para homens
homens = titanic.loc[titanic['sex'] == 'male', 'survived']
homens
```

```python
# Porcentagem de sobrevivncia de homens
totalHomens = homens.count()
(homens.value_counts() / totalHomens) * 100
```

```python
# Filtrando apenas a coluna 'survived' para mulheres
mulheres = titanic.loc[titanic['sex'] == 'female', 'survived']
mulheres
```

```python
# Porcentagem de sobrevivncia de mulheres
totalMulheres = mulheres.count()
(mulheres.value_counts() / totalMulheres) * 100
```

### Crianas x Adultos

Utilizando a mesma abordagem da anlise de sobrevivncia de mulheres e homens, os cdigos abaixo realizam a filtragem por idade, considerando pessoas com 17 anos ou menos como crianas e as demais como adultos. Nota-se que mais da metade das crianas (54,55%) foram salvas, enquanto apenas 39,10% dos adultos sobreviveram. Ainda que mais frgeis, a maior sobrevivncia das crianas talvez seja um indcio da abordagem comum  poca de salvar crianas primeiro.

```python
# Porcentagem de sobrevivncia de crianas e adolescentes (<= 17 anos)
criancas = titanic.loc[titanic['age'] <= 17, 'survived']
totalCriancas = criancas.count()
(criancas.value_counts() / totalCriancas) * 100
```

```python
# Porcentagem de sobrevivncia de adultos (> 17)
adultos = titanic.loc[titanic['age'] > 17, 'survived']
totalAdultos = adultos.count()
(adultos.value_counts() / totalAdultos) * 100
```

### 1 Classe x 2 Classe x 3 Classe

Na comparao das classes, utilizando a mesma abordagem das comparaes anteriores, tambm  possvel identificar um padro. Na primeira classe, 63,68% das pessoas sobreviveram. J a segunda classe obteve uma sobrevivncia de 50,61%. Por fim, na terceira classe, apenas 25,74% dos passageiros sobreviveram. Esses dados apresentam uma clara importncia da classe em relao  sobrevivncia dos passageiros.

```python
# Porcentagem de sobrevivncia na primeira classe
total1classe = titanic.loc[titanic['pclass'] == 1, 'survived'].count()
(titanic.loc[titanic['pclass'] == 1, 'survived'].value_counts() / total1classe) * 100
```

```python
# Porcentagem de sobrevivncia na segunda classe
total2classe = titanic.loc[titanic['pclass'] == 2, 'survived'].count()
(titanic.loc[titanic['pclass'] == 2, 'survived'].value_counts() / total2classe) * 100
```

```python
# Porcentagem de sobrevivncia na terceira classe
total3classe = titanic.loc[titanic['pclass'] == 3, 'survived'].count()
(titanic.loc[titanic['pclass'] == 3, 'survived'].value_counts() / total3classe) * 100
```

### Tabelas Dinmicas ou Piv

Nos exemplos a seguir, so apresentadas anlises semelhantes s anteriores mas, agora, utilizando tabelas piv. No primeiro exemplo, `pivot_class`, observa-se as mdias de sobrevivncia em cada classe (os mesmos valores obtidos anteriormente). J no segundo exemplo, `pivot_sex` a tabela piv apresenta as mdias dos valores de sobrevivncia para cada gnero, indicando, como anteriormente, que 73,97% das mulheres e apenas 21,72% dos homens sobreviveram. Por fim, `pivot_embarked` apresenta que houve uma sobrevivencia maior daqueles que embarcaram em Cherbourg (`C`) com um total de 58,06%.
```python
# Criar uma pivot table para explorar a relao entre 'survived' e outras variveis categricas
pivot_class = pd.pivot_table(titanic, values='survived', index='class', aggfunc='mean')
display(pivot_class)

pivot_sex = pd.pivot_table(titanic, values='survived', index='sex', aggfunc='mean')
display(pivot_sex)

pivot_embarked = pd.pivot_table(titanic, values='survived', index='embarked', aggfunc='mean')
display(pivot_embarked)
```

## Anlise em relao  tarifa paga

Nos grficos a seguir,  possvel observar que a grande maioria dos passageiros pagou uma tarifa menor que 50 dolares. J no segundo grfico, de disperso, nota-se que tarifas mais altas (eixo Y) levaram  maior sobrevivncia (vermelho). Alm disso, as tarifas mais altas esto relacionadas com a classe (tamanho do ponto).

```python
# Distribuio da Tarifa
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], kde=True, color="green")
plt.title('Distribuio da Tarifa Paga pelos Passageiros')
plt.xlabel('Tarifa')
plt.ylabel('Frequncia')
plt.show()
```

No grfico a seguir, a coluna `age`  definida como eixo X e `fare` como eixo Y. A transparncia dos pontos  configurada em 50% (`alpha=0.5`), o tamanho do ponto  configurado para representar a classe (quanto menor melhor), e a cor do ponto representa a sobrevivncia (vermelho indica sobrevivente). Nota-se que tarifas maiores ou idades menores possuem maior aglomerao de sobreviventes.

```python
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5,
            s=titanic['pclass']*20,  # Tamanho do ponto baseado em pclass
            c=titanic['survived'], cmap='coolwarm')  # Cor baseada na sobrevivncia
plt.colorbar(label='Sobreviveu')
plt.title('Relao entre Idade, Tarifa e Sobrevivncia')
plt.xlabel('Idade')
plt.ylabel('Tarifa')
plt.show()
```

## Anlise em Relao  Classe e Sexo dos Passageiros

Os grficos de barras so teis para fazer comparaes entre classes. No exemplo a seguir, estamos utilizando `countplot` para gerar um grfico em barras dos sobreviventes em relao a cada classe. Nota-se que a classe que menos sobreviveu foi a terceira.

```python
# Sobrevivncia por Classe
plt.figure(figsize=(8, 6))
sns.countplot(x='class', hue='survived', data=titanic, palette='pastel')
plt.title('Sobrevivncia por Classe de Passageiro')
plt.xlabel('Classe')
plt.ylabel('Nmero de Passageiros')
plt.legend(title='Sobreviveu', labels=['No', 'Sim'])
plt.show()
```

Outro exemplo de uso de `countplot`  dado a seguir. Neste caso, cada classe de sobrevivncia (Sim ou No) tambm  representada pelas cores e, no eixo X, h a separao por gnero. Nota-se que homens sobreviveram bem menos que mulheres.

```python
# Sobrevivncia por Sexo
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='survived', data=titanic, palette='pastel')
plt.title('Sobrevivncia por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Nmero de Passageiros')
plt.legend(title='Sobreviveu', labels=['No', 'Sim'])
plt.show()
```

No exemplo a seguir, utilizamos `subplots` para agrupar trs grficos de barras (`barplot`) em disposio 1x3 (uma linha e 3 colunas). A varivel `axes` recebe um *array* com os trs espaos a serem utilizados pelos graficos em seguida (utilizando o parametro `ax`). O primeiro apresenta a taxa de sobrevivncia por classe, no segundo tem-se a taxa por sexo e, por fim, a taxa por ponto de embarque. Esse tipo de grfico apresenta o desvio padro por meio de uma linha na vertical em cada barra. Com isso,  possvel observar, por exemplo, que o desvio padro para o ponto de embarque em Queenstown foi maior que o desvio padro do ponto em Southampton.

```python
# Anlise de propores de sobrevivncia por classe, sexo e porto de embarque
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Sobrevivncia por Classe
sns.barplot(x='class', y='survived', data=titanic, ax=axes[0])
axes[0].set_title('Taxa de Sobrevivncia por Classe de Passageiro')
axes[0].set_ylabel('Taxa de Sobrevivncia')
axes[0].set_xlabel('Classe')

# Sobrevivncia por Sexo
sns.barplot(x='sex', y='survived', data=titanic, ax=axes[1])
axes[1].set_title('Taxa de Sobrevivncia por Sexo')
axes[1].set_ylabel('Taxa de Sobrevivncia')
axes[1].set_xlabel('Sexo')

# Sobrevivncia por Porto de Embarque
sns.barplot(x='embark_town', y='survived', data=titanic, ax=axes[2])
axes[2].set_title('Taxa de Sobrevivncia por Porto de Embarque')
axes[2].set_ylabel('Taxa de Sobrevivncia')
axes[2].set_xlabel('Porto de Embarque')

plt.tight_layout()
plt.show()
```

## Visualizar Distribuio dos Dados (Atributos Numricos)

## Visualizar a Presena de Outliers (Atributos Numricos)

*Outliers* so valores que diferem significativamente do resto dos dados em um conjunto de observaes. Eles podem ocorrer devido a erros de medio, variaes genunas nos dados ou falhas no processo de coleta de dados. Identificar e entender *outliers*  crucial porque eles podem distorcer anlises estatsticas, como mdias e regresses, levando a interpretaes errneas ou concluses enganosas. Em contextos prticos, analisar *outliers* ajuda a aprimorar os modelos, garantindo que eles sejam robustos e representativos da realidade observada, ou mesmo revelar *insights* valiosos sobre comportamentos atpicos ou casos extremos.

De maneira similar ao exemplo anterior, no cdigo a seguir, vrios graficos so organizados em uma nica imagem utilizando o mtodo `subplots`. Neste caso, so criados grficos do tipo `boxplot` para cada coluna numrica. O parmetro `orient='vertical'` define a orientao dos grficos.

```python
fig, ax = plt.subplots(ncols=len(numericos.columns), nrows=1, figsize=(25, 5))
plt.suptitle("Boxplots dos Atributos Numricos")

# Grfico para cada atributo numrico
for i in range(0, len(numericos.columns)):
    feature = numericos.columns[i]
    sns.boxplot(numericos[feature], ax=ax[i], orient='vertical')
```
